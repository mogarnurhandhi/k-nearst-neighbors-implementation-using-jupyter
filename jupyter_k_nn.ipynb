```python
#melakukan import library python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#melakukan pemanggilan dataset yang akan dilakukan
dataset = pd.read_csv('Social Network Ads.csv')
```

```python
#menampilkan data 5 teratas
dataset.head()
```

```python
#preposesing
#melihat data apakah ada yang kosong
dataset.isna().sum()
```

```python
#preposesing
#membersihkan missing value dengan drop
dataset.dropna(inplace=True)
```

```python
#preposesing
#melihatkembali  data apakah ada yang kosong
dataset.isna().sum()
```

```python
#membagi data kedalam x dan y
x = dataset.iloc[:,[2,3]].values
#semua baris yang diambil pada kolom 2 dan 3
y = dataset.iloc[:,-1].values
#semua baris dan kolom terakhir
#x untuk reguler atributnya dan y untuk labelnya
```

```python
#enampilkan data x dan y
print(x)
print(y)
```

```python
#membagi data training dan data testing
#menggunakan model selection denganfunction train to split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)
#porsi 25% untuk testing
#jumlah data keseluruhan 400an tetapi telah dikurangi oleh proses drop data yang missing
#sehingga data testing kurang lebih 100 dan training 300
#data tersebut bisa kurang
```

```python
#menampilkan data x_train dan melihat jumlahnya
print(x_train)
len(x_train)
```

```python
#menampilkan data x
len(x)
```

```python
#menampilkan data x_test dan melihat jumlahnya
print(x_test)
len(x_test)
```

```python
#menampilkan data x
len(y)
```

```python
#menampilkan jumlah data y_test
len(y_test)
```

```python
#menampilkan jumlah data y_train
len(y_train)
```

```python
#proses scaling dengan standart scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

#yang di scale hanya x keran jarak age dan estimated salary scalenya terlalu besar
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
#menampilkan data yang sudah di skala
#umur dan estimate salary
print(x_train)
```

```python
print(x_test)
```

```python
#menggunakan algoritma klasifikasi k nn
from sklearn.neighbors import KNeighborsClassifier
#k ditentukan 5
#mestricnya bisa minkowski atau eclidean
classifier = KNeighborsClassifier(n_neighbors= 5, metric='minkowski', p=2)
classifier.fit(x_train, y_train)
```

```python
#menentukan prediksi
y_pred = classifier.predict(x_test)
```

```python
#menghitung evaluasi atau validasi menggunakan confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
#prediksi benar dan kenyataan benar 52
#prediksi salah dan kenyataan salah 29
#niali akurasi 52+29 / (52+7+9+29) * 100%
#0.835052
#model yang dihasilkan sudah bagus
```

```python
#hasil akurasi
np.mean(y_pred == y_test)
```

```python
#visualisasi distribusi datanya dengan k nn data training
#menggunakan matplotlib
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train 
#bagi data training
#bagi variable x nya
#membuat mesh dengan meshgrid dan dibagi x1 untuk umur x2 untuk estimasi gaji
#range umur start nilai minimal sampai maksimal sampai stepnya 0.01
#range estimasi gaji start nilai minimal sampai maksimal sampai stepnya 0.01
x1, x2 = np.meshgrid(np.arange(start=x_set[:,0].min()-1, stop=x_set[:,0].max()+1, step=0.01),
                     np.arange(start=x_set[:,1].min()-1, stop=x_set[:,1].max()+1, step=0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red','green')))

#membuat range dari minimal ke maksimal
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
#perulanganya menggunakan for
for i, j in enumerate (np.unique(y_set)):
    #menscatterkan dengan titiknya
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
    c = ListedColormap(('red','green'))(i), label =j)

#membuat judulnya
plt.title('Klasifikasi Data dengan K-NN (Data Training)')
plt.xlabel('Umur')
plt.ylabel('Estimasi Gaji')
plt.show()
```

```python
#visualisasi distribusi datanya dengan k nn data testing
#menggunakan matplotlib
from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test 
#bagi data training
#bagi variable x nya
#membuat mesh dengan meshgrid dan dibagi x1 untuk umur x2 untuk estimasi gaji
#range umur start nilai minimal sampai maksimal sampai stepnya 0.01
#range estimasi gaji start nilai minimal sampai maksimal sampai stepnya 0.01
x1, x2 = np.meshgrid(np.arange(start=x_set[:,0].min()-1, stop=x_set[:,0].max()+1, step=0.01),
                     np.arange(start=x_set[:,1].min()-1, stop=x_set[:,1].max()+1, step=0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red','green')))

#membuat range dari minimal ke maksimal
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
#perulanganya menggunakan for
for i, j in enumerate (np.unique(y_set)):
    #menscatterkan dengan titiknya
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
    c = ListedColormap(('red','green'))(i), label =j)

#membuat judulnya
plt.title('Klasifikasi Data dengan K-NN (Data Testing)')
plt.xlabel('Umur')
plt.ylabel('Estimasi Gaji')
plt.show()
```
